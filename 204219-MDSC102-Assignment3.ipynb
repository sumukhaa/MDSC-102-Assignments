**Importing required packages**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

**Binomial distribution**

```python
def binomial(n,p,x):
    #Probability density function of Binomial distribution is
    return (np.math.comb(n,x)) * (p**x) * ((1-p)**(n-x))

n = 50
p = 0.5

x_values = np.arange(0,n+1)
bin_probs = [binomial(n,p,x) for x in x_values]

plt.bar(x_values,bin_probs)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution')
plt.show()
```

**Poisson distribution**

```python
def poisson(l,x):
    e=np.exp(1)
    #pdf of poisson distribution
    return ((e**(-l)) * (l**(x)))/ np.math.factorial(x)

l=0.01
l=1
l=5

x_values = np.arange(0,20)
poisson_probs = [poisson(l,x) for x in x_values]

plt.bar(x_values,poisson_probs)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.show()
```

**Exponential distribution**

```python
def exponential(l,x):
    e=np.exp(1)
    #pdf of exponential distribution
    return l*(e**((-l)*x))


#small step taken to get a curve like figure
x_values = np.arange(0,15,0.2)
exp_probs = np.array([exponential(1,x) for x in x_values])  # l=1

plt.plot(x_values,exp_probs)
plt.xlabel('Continuous r.v')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution')
plt.show()
```

**Normal distribution**

```python
#m=mean v=variance
def normal(x,m,v):
    e=np.exp(1)
    #pdf of normal distribution
    return (e**((-0.5)*((x-m)**2))/v)/np.sqrt(2*v*np.pi)

m = 0
v = 4

#small step taken to get a curve like figure
x_values = np.arange(-15,15,0.2)
normal_probs = np.array([normal(x,m,v) for x in x_values])

plt.plot(x_values,normal_probs)
plt.xlabel('Continuous r.v')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.show()
```

**Gamma distribution**

```python
#a=alpha l=lambda
def gamma(a,l,x):
    e=np.exp(1)
    return (x**(a-1))*(e**(-x))*(l**a)/np.math.factorial(a)

a = 5
l = 4

x_values = np.arange(1,30,0.1)
gamma_probs = np.array([gamma(a,l,x) for x in x_values])

plt.plot(x_values,gamma_probs)
plt.xlabel('Continuous r.v')
plt.ylabel('Probability Density')
plt.title('Gamma Distribution')
plt.show()
```
