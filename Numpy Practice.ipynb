**Defining a function**

```python
def func(i,j):
    return 2*i*j
```

```python
np.fromfunction(func,(4,3))
```

**Using lambda to define the above function and call it in single line**

```python
np.fromfunction(lambda i,j: 2*i*j,(4,3))
```

```python
import matplotlib.pyplot as plt

f = lambda x,y: 0.3*x**3 + y**2 + 2*x*y - 6*x - 3*y + 4
x = np.arange(-15,15,0.1)
y = np.arange(-15,15,0.1)

fig = plt.figure(figsize = (16,8))
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
N=50

cp = plt.contour(X,Y,Z,N, cmap='jet')
plt.clabel(cp, inline=1, fontsize=8)

plt.axis('scaled')
plt.show()
```

```python
import matplotlib.pyplot as plt

f = lambda x,y: x**2 - 2*x*y + 4*y**2
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

fig = plt.figure(figsize = (8,16))
X,Y = np.meshgrid(x,y)
Z = f(X,Y)

levels = np.arange(-1.0,50,4)

cp = plt.contour(X,Y,Z, cmap='jet', linestyles= 'dashed', levels=levels)
plt.clabel(cp, inline=1, fontsize=8)

plt.axis('scaled')
plt.show()
```
