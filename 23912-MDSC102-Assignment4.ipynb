```python
i=1
plt.figure(figsize=(25,30))
for feature in df:
    if feature!='Class' and feature!='Alcohol':
        plt.subplot(4,3,i)
        sns.scatterplot(x=df[feature], y=df['Alcohol'], hue=df['Class'])
        plt.grid()
        i+=1
```
